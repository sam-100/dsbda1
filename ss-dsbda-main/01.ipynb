```python
df['Age']=df['Age'].astype(int)
df['Number']=df['Number'].astype(int)
```

```python
df.dtypes
```

```python
std = df['Salary'].std()
mean = df['Salary'].mean()
mini = df['Salary'].min()
maxi = df['Salary'].max()
mean,std, mini, maxi
```

```python
# z score normalization
df['Salary_z']=(df['Salary']-mean)/std
```

```python
df
```

```python
# mini-max normalization
df['Salary_mm']=(df['Salary']-mini)/(maxi-mini)
```

```python
df
```

```python
s=df['Salary_z'].skew()
s
```

```python
df['Salary_mm'].hist()
```

```python
sns.distplot(df['Salary_mm'],hist=True)
```

```python
df['Salary_cbrt']=df['Salary']**(1/3)
df['Salary_cbrt'].skew()
#sns.distplot(df['Salary_cbrt'],hist=True)
```

```python
df['Salary_log']=np.log(df['Salary'])
df['Salary_log'].skew()
#sns.distplot(df['Salary_log'],hist=True)
```
