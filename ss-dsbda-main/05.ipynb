# Binning

```python
#to check ages (for reference to create bins)
df.Age.value_counts().sort_index()
```

```python
#we can make bins like 19-24, 25-30, 31-36 etc. for creating bin

bin_value=[19,25,31,36,40]
label_value=['19-24','25-30','31-35','36-40']
df['AgeGroup']=pd.cut(df['Age'], bins=bin_value, labels=label_value, right=False)
```

```python
df
```

# Grouping

```python
#summary statistics of each group

var=df.groupby('AgeGroup')['Salary']
mean=var.mean()
median=var.median()
describe=var.describe()
print('Average Salary by mean:\n', mean, median, describe)
```
